```python
# Return: keyword send back all the values to the caller
# return is similar to print(), when we use the return call the function with print()

def sum(*num):
    total = 0
    for i in num:
        total = total + i
    return total
print(sum(10,20,10,10))
```

```python
#write a function to calculate the factorial of a number --> 5*4*3*2*1
fact=1
for i in range(1,5):
    fact=fact*i
   # print(fact)
    
#using def function factorials.    
def factorial(n,m):
    fact=1
    for i in range(n,m+1):
        fact=fact*i
    return fact
print(factorial(1,5))



        
```
