```python
# Encapsulation means wrapping data (variables) and methods into a single unit (class) and controlling 
# access to that data

```

```python
# Public:
# Members that can be accessed from anywhere.
class Student:
    def __init__(self, name):
        self.name = name   # public variable

    def display(self):
        print(self.name)

s = Student("Khan")
s.display()
# print(s.name)

```

```python
# Protected
# Members that should be accessed only inside the class and its child classes.
# (Uses single underscore _)
class Student:
    def __init__(self, marks):
        self._marks = marks   # protected variable

    def show(self):
        print(self._marks)

class Result(Student):
    def display(self):
        print(self._marks)

r = Result(85)
r.display()
```

```python
# Private:
# Members that can be accessed only inside the same class.
# (Uses double underscore __)

class Student:
    def __init__(self):
        self.__fees = 5000   # private variable

    def show_fees(self):
        print(self.__fees)

s = Student()
s.show_fees()
```
