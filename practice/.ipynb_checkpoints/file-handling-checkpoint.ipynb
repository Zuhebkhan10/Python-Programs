```python
# x-mode() :Create the text-file
f=open("Got.txt","x")
print(f)
f.close()
```

```python
# w-mode() :write the file
f=open("Got.txt","w")
f.write("Winter is coming This is john snow")
print(f)
f.close()
```

```python
# a-mode(): add the data
f=open("Got.txt","a")
f.write("Once upon a timejohn snow because sansa and arya is wonderful girls")
print(f)
f.close()
```
