```python
"""Regular expression is sequence of character that defines search patterns for string searching
matching validatioand manipulation"""

"""
RegEx Functions
There module offers a set of functions that allows us to search a string for a match:

findall: Returns a list containing all matches
search: Returns a Match object if there is a match anywhere in the string
split:	Returns a list where the string has been split at each match
sub:   Replaces one or many matches with a string """
```

```python
# Search() search the book in yes or no
import re
m="one chapter in a book"
res=re.search("book",m)
print(res)

```

```python
# Findall()  find the book present or not this program 
import re
lib="the full book with many chapter"
res=re.findall("book",lib)
if (res):
    print("Yes Present")
else:
    print("Not present")
print(res)
```

```python
# replace() sub() replace the max into Eleven
import re 
h="""Max is so pretty girl but she is entire life for depression Max and the character is so amazing 
perform the all episode and her favorite song is kate bush many to listening."""
res=re.sub("Max","Eleven",h)
print(res)
```

```python
# Split() all white-space printing
import re 
char="Veccna said :Small steps every day lead to big success."
res=re.split(" ",char)
print(res)
```

```python
#Replace all white-space characters with the digit "2"
import re
s="call me khan"
r=re.sub("\s","2",s)
print(r)
```

```python
# Match Object
# A Match Object is an object containing information about the search and the result.
'''
.span() returns a tuple containing the start-, and end positions of the match.
.string returns the string passed into the function
.group() returns the part of the string where there was a match
'''
```

```python
#Search for an lower case "f" character in the beginning of a word, and print its position
fav="Biryani is my favorite food "
r=re.search(r"\bf\w+",fav)
print(r.span())
```

```python
# Print the string passed into the function:
fav="Biryani is my favorite food "
r=re.search(r"\bf\w+",fav)
print(r.string)
```

```python
# Print the part of the string where there was a match.
fav="Biryani is my favorite food "
r=re.search(r"\bf\w+",fav)
print(r.group())
```

```python
"""
cap symbols(^) --> Start with
plus(+) --> add 1 or more
Dollar-sign($) --> ends withs
Digit shorthand(\d) --> digits(0-9)
word character shorthand(\w) --> wordCharacter(a-z and A-Z ,0-9)
whitespace shorthand(\s) --> space
Asterisk(*) --> 0 or more repitations
character class[]--> set of character
Dot. --> any character expect newline
question mark ? --> 0 or 1 repitations
 """
```

```python
# cap symbol ^
import re
txt="hello duniya"
pattern="^hello"

res=re.search(pattern,txt)
print(res)
```

```python
# plus + 1 or more repetitions
import re
txt="duniya dost darr"
pattern="d+"

res=re.findall(pattern,txt)
print(res)
```

```python
# Dollar $ → Ends with
import re
txt="the price is $300"

res=re.search(r"\$\d+",txt)
print(res)
```

```python
# Digit shorthand \d → Digits (0–9)
import re
txt="My age is 21 years old"
pattern="\d"

res=re.findall(pattern,txt)
print(res)
```

```python
# 5️⃣ Word character \w → a-z, A-Z, 0-9, _
txt="python_DA"
pattern="\w"
res=re.findall(pattern,txt)
print(res)
```

```python
# Whitespace \s → Space
import re
txt="Call me khan"
pattern="\s"

res=re.findall(pattern,txt)
print(res)
```

```python
# Asterisk * → 0 or more repetitions
import re
txt="abbaa"
pattern="ab*"

res=re.findall(pattern,txt)
print(res)
```

```python
# Character class [] → Set of characters
import re
n="pray trust wait"
p="[t]"

res=re.findall(p,n)
print(res)
```

```python
# Dot . → Any character except newline
import re
n="rat bat cat late "
p=".at"

res=re.findall(p,n)
print(res)
```

```python
#  Question mark ? → 0 or 1 repetition
txt="color colour"
pattern="colou?r"

r=re.findall(pattern,txt)
print(r)
```
