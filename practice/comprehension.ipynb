```python
# list-comprehensions:

# [expression for item in iterable if condition]
a=[10,22,11,12,33,39,90]

even=[i for i in a if i%2==0]
odd=[i for i in a if i%2==1]

print(even)
print(odd)
```

```python
# Creating a list of squares
# s=[2,3,4,5,10,11,12]
# square=[i for i in s if i**2]
# print(square)

s=[num**2 for num in range(1,6)]
print(s)
```

```python
# 2. Dictionary comprehension
# {key_expression: value_expression for item in iterable if condition}

res={num:num**2 for num in range(2,6)}
print(res)
```

```python
# Mapping states to capitals

a=["Uttar Pradesh","Telangana","Karnataka","West Bengal"]
b=["Lucknow","Hyderabad","Benguluru","Kolkata"]

res={state:capital for state,capital in zip(a,b)}
print(res)
```
